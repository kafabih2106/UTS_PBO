```
import pandas as pd
import matplotlib.pyplot as plt

# Baca file + delimiter ;
df = pd.read_csv("Percobaan_mhs.csv (1).csv", sep=';')

# Cek data
df.head()
```

```
plt.figure(figsize=(8,6))

scatter = plt.scatter(
    df['X1'],          # sumbu X
    df['X2'],          # sumbu Y
    c=df['X3'],        # warna
    s=df['X4'] * 20,   # ukuran titik
    cmap='viridis',
    alpha=0.7
)

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Scatter Plot 4 Variabel (X1, X2, X3, X4)')
plt.colorbar(scatter, label='X3')
plt.grid(True)
plt.show()
```

```
plt.figure(figsize=(8,6))

plt.scatter(
    df['X1'],
    df['X2'],
    c=df['X3'],
    s=df['X4'] * 20,
    cmap='viridis',
    alpha=0.7
)

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Scatter Plot 4 Variabel')
plt.colorbar(label='X3')
plt.grid(True)

plt.savefig("scatterplot_4_variabel.png", dpi=300)
plt.show()
```
