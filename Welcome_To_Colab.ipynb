```
import matplotlib.pyplot as plt
import random

```

```
# --- DATA UNTUK SOAL 1 ---
# Minimal 5 Kategori
kategori = ['Transportasi', 'Akomodasi', 'Makanan', 'Hiburan', 'Belanja', 'Lain-lain'] # 6 Kategori (>= 5)
# Nilai Acak (15-85)
nilai_figure = [random.randint(15, 85) for _ in range(len(kategori))]
```

```
# --- VISUALISASI FIGURE ---
print("Tugas 1: Visualisasi Figure")
```

```
# Atur Ukuran Figure (10, 6)
plt.figure(figsize=(10, 6))
```

```
# Buat Bar Chart dengan Warna Hijau
plt.bar(kategori, nilai_figure, color='green')
```

```
# Tambahkan Judul dan Label
plt.title('1. Pengeluaran Bulanan Berdasarkan Kategori', fontsize=16)
plt.xlabel('Kategori', fontsize=12)
plt.ylabel('Jumlah (Juta Rupiah)', fontsize=12)
plt.grid(axis='y', alpha=0.6, linestyle='--')
plt.xticks(rotation=15) # Agar label x tidak bertumpuk
```

```
import matplotlib.pyplot as plt
import numpy as np
import random

print("Tugas 2: Visualisasi 4 Axes dalam 1 Figure (Kode Diperbaiki)")

# --- DATA UNTUK SOAL 2 ---
# Nilai dibuat berbeda-beda untuk memenuhi ketentuan "Nilai tidak boleh sama dengan contoh"

# Data 1 (Line Plot): Nilai float kecil (0-25)
x_data = np.linspace(0, 5, 20)
y1 = x_data ** 3

# Data 2 (Scatter Plot): Nilai sedang (10-40)
# Menggunakan numpy array agar konsisten, tetapi bisa juga list biasa
x2 = np.array([random.uniform(10, 40) for _ in range(15)])
y2 = np.array([random.uniform(10, 40) for _ in range(15)])

# Data 3 (Bar Plot): Nilai besar (100-800)
labels_bar = ['Q1', 'Q2', 'Q3', 'Q4']
nilai_bar = [random.randint(100, 800) for _ in range(len(labels_bar))]

# Data 4 (Histogram): Nilai sangat besar (~500)
data_hist = np.random.normal(loc=500, scale=30, size=50)


# --- VISUALISASI AXES ---

# Buat 1 Figure yang terdiri dari 4 Axes (2 baris, 2 kolom)
# figsize diperkecil sedikit agar lebih aman
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('2. Analisis Data Multi-Variabel (4 Axes)', fontsize=16)


# Axes 1 (Baris 0, Kolom 0): Line Plot
axes[0, 0].plot(x_data, y1, color='red')
axes[0, 0].set_title('A. Tren Pertumbuhan (Line)')

# Axes 2 (Baris 0, Kolom 1): Scatter Plot
axes[0, 1].scatter(x2, y2, color='blue', marker='o')
axes[0, 1].set_title('B. Korelasi Acak (Scatter)')

# Axes 3 (Baris 1, Kolom 0): Bar Plot
axes[1, 0].bar(labels_bar, nilai_bar, color='orange')
axes[1, 0].set_title('C. Pendapatan Kuartalan (Bar)')

# Axes 4 (Baris 1, Kolom 1): Histogram
axes[1, 1].hist(data_hist, bins=12, color='purple', edgecolor='black')
axes[1, 1].set_title('D. Distribusi Data (Histogram)')

# Merapikan layout (menggunakan metode yang lebih standar)
plt.tight_layout()

# Sesuaikan posisi judul agar tidak terpotong setelah tight_layout
fig.subplots_adjust(top=0.92)

plt.show() # Tampilkan plot 2
```
