```
# Data
kategori = ['A', 'B', 'C', 'D', 'E']
nilai = [12, 25, 18, 30, 22]
```

```
# Membuat figure
plt.figure(figsize=(10, 6))
plt.bar(kategori, nilai, color='green')
```

```
# Menambahkan judul dan label
plt.title('Visualisasi Data - Warna Hijau', fontsize=14)
plt.xlabel('Kategori')
plt.ylabel('Nilai')
```

```
# Menampilkan grafik
plt.show()
```

```
# Nomor 2 - Visualisasi Data menggunakan 4 Axes
import matplotlib.pyplot as plt
```

```
# Membuat figure
fig, axs = plt.subplots(2, 2, figsize=(10, 6))  # 4 axes (2x2 grid)
```

```
# Data untuk setiap axes (semua berbeda)
data1 = [5, 10, 15, 20, 25]
data2 = [3, 8, 12, 19, 24]
data3 = [7, 14, 10, 18, 22]
data4 = [4, 9, 13, 17, 21]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```
# Plot setiap axes
axs[0, 0].bar(kategori, data1, color='green')
axs[0, 0].set_title('Grafik 1')
```

```

axs[0, 1].plot(kategori, data2, marker='o', color='green')
axs[0, 1].set_title('Grafik 2')
```

```
axs[1, 0].barh(kategori, data3, color='green')
axs[1, 0].set_title('Grafik 3')
```

```
axs[1, 1].scatter(kategori, data4, color='green')
axs[1, 1].set_title('Grafik 4')
```

```
# Menyesuaikan layout
plt.tight_layout()
plt.show()
```
